```python
#Given a string, convert it to uppercase and lowercase.
name = "Sowmya"
print(name.upper()) #prints uppercase
print(name.lower()) #prints lowercase
```

```python
#Write a program to count the number of vowels in a string.
vowels = ('a', 'e', 'i', 'o', 'u')
count2 = 0
name2 = "Sowmya"

for ch in name2.lower():   # convert to lowercase to handle capital letters
    if ch in vowels:
        count2 += 1

print("Number of vowels:", count2)
```

```python
#Check whether a given string is a palindrome (example: madam).
pal = "Madam"

pal = pal.lower()
if pal == pal[::-1]:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
#Given a sentence, count the number of words in it.
sen = " Hi! My name is Sowmya!"
split_sen = sen.split()
print(len(split_sen))
```

```python
#Replace all spaces in a string with underscores (_).
sen = " Hi! My name is Sowmya!"
print(sen.replace(" ", "_"))
```

```python
#Write a program to reverse a string without using built-in reverse functions.
name3 = "Sowmya"
rev = ""

for ch in name3:
    rev = ch + rev

print(rev)
```

```python
#Write a program to reverse a string without using built-in reverse functions.
name4 = "Sowmya"
rev1 = ""

for i in range(len(name4) - 1, -1, -1):
    rev1 += name4[i]

print(rev1)
```

```python
''' String "   Learning Python is simple and Python is powerful   "
    1.	Given a string, remove leading and trailing spaces.
	2.	Check whether a string starts with a given character.
	3.	Find the index of the first occurrence of a character in a string.
	4.	Count how many times a specific character appears in a string.
	5.	Split a sentence into a list of words.
	6.	Join a list of words into a single string with spaces.
	7.	Replace a word in a sentence with another word
'''
user_input = "   Learning Python is simple and Python is powerful   "
# Given a string, remove leading and trailing spaces.
cleaned_input = user_input.strip()
print(cleaned_input)

# Check whether a string starts with a given character.
print(cleaned_input.startswith('L'))

# Find the index of the first occurrence of a character in a string.
index = cleaned_input.find('P')
print(index)
#prints index if found else it returns -1

# Count how many times a specific character appears in a string.
count_o = cleaned_input.count('o')
print(count_o)

# Split a sentence into a list of words.
words = cleaned_input.split()
print(words)
#Split takes string as input and gives list as output

#	Join a list of words into a single string with spaces.
joined_sentence = " ".join(words)
print(joined_sentence)

#	Replace a word in a sentence with another word
replaced_text = cleaned_input.replace("Python","Javascript")
print(replaced_text)
```

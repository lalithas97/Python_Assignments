```python
# Question 41: Check if string contains balanced parentheses: "((()))"
print("\nQuestion 41: Check if string contains balanced parentheses: '((()))'")
# Your code here

text = "((()))"
balance = 0
is_balanced = True

for ch in text:
    if ch == '(':
        balance += 1
    elif ch == ')':
        balance -= 1

    if balance < 0:
        is_balanced = False
        break

if balance != 0:
    is_balanced = False

if is_balanced:
    print("Balanced parentheses")
else:
    print("Not balanced")
```

```python
# Question 46: Convert "hello world" to Pig Latin
print("\nQuestion 46: Convert 'hello world' to Pig Latin")
# Your code here
'''
For each word:
1️⃣ If a word starts with a vowel → add "way" at the end
2️⃣ If a word starts with a consonant →
    • move the first letter to the end
    • add "ay"
'''
sentence = 'hello world'
arr = sentence.split()
result = []

vowels = "aeiou"

for word in arr:
    if word[0].lower() in vowels:
        result.append(word + "way")
    else:
        result.append(word[1:] + word[0] + "ay")

print(" ".join(result))
```

```python
# Question 47: Check if string is a valid IPv4 address: "192.168.1.1"
print("\nQuestion 47: Check if string is a valid IPv4 address: '192.168.1.1'")
# Your code here
# IPv4 address has 4 parts when seperated by dots and the those 4 parts contain only digits that has value from 0 to 255(inclusive)
address = "192.168.1.1"
cleaned_address = address.split(".")

flag = True   # assume valid first

if len(cleaned_address) != 4:
    flag = False
else:
    for word in cleaned_address:
        if not word.isdigit():
            flag = False
            break
        elif not (0 <= int(word) <= 255):
            flag = False
            break

print("Valid IPv4 address" if flag else "Invalid IPv4 address")
```

```python
# Question 49: Convert "hello world" to ROT13 encoding
print("\nQuestion 49: Convert 'hello world' to ROT13 encoding")
# Your code here
text = "hello world"
result = ""

for ch in text:
    if ch.isalpha():
        # find position (0–25)
        offset = ord(ch) - ord('a')
        # rotate by 13
        new_char = chr((offset + 13) % 26 + ord('a'))
        result += new_char
    else:
        result += ch   # keep spaces as-is

print(result)
```

```python
# Question 50: Check if string is a valid credit card number: "4532015112830366"
print("\nQuestion 50: Check if string is a valid credit card number: '4532015112830366'")
# Your code here

# Step1:
# 1) Should a credit card contain only digits?
# 2) Does it have a fixed length (usually 16 digits)?

# Step2:
'''
  Starting from the rightmost digit:

    Move right to left

    Double every second digit

    If doubling gives a number > 9, subtract 9

    Add all digits together

    If the final sum is divisible by 10, the card is valid
'''

card_number = "4532015112830366"

# Step 1: Basic checks
if len(card_number) != 16 or not card_number.isdigit():
    print("Invalid credit card number")
else:
    total = 0
    i = 0

    # Step 2: Process digits from right to left
    for ch in card_number[::-1]:
        digit = int(ch)

        # Step 3: Double every second digit
        if i % 2 == 1:
            digit *= 2
            if digit > 9:
                digit -= 9

        total += digit
        i += 1

    # Step 4: Final check
    if total % 10 == 0:
        print("Valid credit card number")
    else:
        print("Invalid credit card number")
```
